```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("fivethirtyeight")

%matplotlib inline
%config InlineBackend.figure_format = "svg"
```

```python
dataset = pd.read_csv("./datasets/creditcard.csv")
dataset
```

```python
class_counts = dataset.value_counts("Class")
print(class_counts)

class_counts.plot(kind="bar", logy=True)
```

```python
normal_trans_v1 = dataset.query('Class == 0')["V1"]
scam_trans_v1 = dataset.query('Class == 1')["V1"]

plt.hist(scam_trans_v1.values, bins=20, alpha=0.5, color="red", density=True, label="Class 1")
plt.hist(normal_trans_v1.values, bins=20, alpha=0.5, color="gray", density=True, label="Class 0")
plt.xlabel("V1")
plt.legend()
```
